```python
import networkx as nx
import numpy as np
import h5py
import gdy
from collections import Counter
import cPickle as pickle
```

```python
def edgedata(g):
    """
    Return edge data in a form that can be output as HDF5
    For the purposes of reconstituting graphs in Julia
    """
    e = np.array(g.edges()) + 1
    return e

def max_shortest_path_length(G):
    """This is pretty slow"""
    a = nx.shortest_path_length(G)
    l = []
    for key in a:
        l.extend(a[key].values())
    return max(l)

def mean_shortest_path_length(G):
    """This is pretty slow"""
    a = nx.shortest_path_length(G)
    l = []
    for key in a:
        l.extend(a[key].values())
    return np.mean(l)
```

# Generate Configuration File

```python
G1 = nx.barabasi_albert_graph(100,3,0)
G2 = nx.configuration_model([4]*50+[8]*50, seed=0)
G2.remove_edges_from(G2.selfloop_edges())
G2 = nx.Graph(G2)
edata1 = edgedata(G1)
edata2 = edgedata(G2)
```

```python
# Other parameters
otimes = np.array([  1.,   2.,   5.,  10.,  20.,  50.])
r0s = np.linspace(.75, 1.5, 11)
alphas = np.logspace(-2, 2, 11)
kmean1 = mean(G1.degree().values())
kmean2 = mean(G2.degree().values())
nruns = 100
filename1 = u'2016_reunion_tour/config_ba';
filename2 = u'2016_reunion_tour/config_cm';
desc1 = u'100 node BA graph with m=3, seed=0';
desc2 = u'100 node configuration model graph, k1=4, k2 = 8, p1 = .5; seed=0';
```

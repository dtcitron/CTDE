```python
# define a function for reading in the .txt file as a 3d Array
def readtxt(filename):
    f = open(filename, 'r')
    full = []
    data = []
    for line in f:
        if len(line) == 1:
            full.append(data)
            data = []
        else:
            l = line.rstrip('\n').split(',')
            l = map(int, l)
            # print l
            data.append(l)
    full.append(data)
    return full
```

```python
z = array(readtxt('z.txt'))
```

```python
# Count the number of trajectories that have failed at each time
# z[:, :, 1] Returns the number of susceptibles at each time point for all trajectories
np.array([sum((z[i, :, 1] > 0)) for i in range(len(otimes))])
```

```python
def countliving(z, otimes):
    """
    Count the number of remaining 'living' trajectories vs time
    z : matrix of trajectory ensemble
    otimes : the list of observation times
    returns list of numbers of living trajectories
    """
    return np.array([np.sum((z[i, :, 1] > 0)) for i in range(len(otimes))])
```

```python
c = countliving(z, otimes);
plt.plot(otimes, c, marker = 'o', color = 'k')
xlabel('time');
ylabel('number of trajectories');
```

```python
# When computing statistics on the QS distribution, need to exclude all extinct trajectories

def trajmeans(z, otimes, mode = 1):
    """
    Exclude all extinct trajectories from statistics calculations
    z : matrix of trajectory ensemble
    otimes : the list of observation times
    mode : pick S, I, or R trajectories; defaults to 1 to show infecteds
    returns list of means, excluding extinct trajectories
    """
    return np.array([np.mean(z[i,:,mode][np.where(z[i, :, mode] > 0)]) for i in range(len(otimes))])

def trajstds(z, otimes, mode = 1):
    """
    Exclude all extinct trajectories from statistics calculations
    z : matrix of trajectory ensemble
    otimes : the list of observation times
    mode : pick S, I, or R trajectories; defaults to 1 to show infecteds
    returns list of standard deviations, excluding extinct trajectories
    """
    return np.array([np.std(z[i,:,mode][np.where(z[i, :, mode] > 0)]) for i in range(len(otimes))])
```

```python
m = trajmeans(z, otimes, 1);
s = trajstds(z, otimes, 1);
print m
print s
plot(otimes, m, marker = 'o', color = 'k')
plot(otimes, m+s, marker = 'o', color = 'r')
plot(otimes, m-s, marker = 'o', color = 'r')
xlabel('Time')
ylabel('Number infected')
ylim(0,100)
```
